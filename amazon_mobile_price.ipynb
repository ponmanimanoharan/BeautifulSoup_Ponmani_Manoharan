```python
url_src.insert(0, "https://www.amazon.in/s?k=mobile+phones+under+20%2C000&rh=p_36%3A-2000000&qid=1620446282&rnid=1318502031&ref=sr_pg_1")
```

```python
new_lst = []
def url_scrap(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    a = soup.find_all("img")
    ex = a[3:]
    for x in ex:
        c = x.attrs["src"]
        new_lst.append(c)
    print(new_lst)
    
    
    
```

```python
for x in url_src:
    images = url_scrap(x)
    print(images)
    
```

```python
new_lst
```

```python
mobile_name = []
def name(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    name = soup.find_all("span" , class_="a-size-medium a-color-base a-text-normal")
    for x in name:
        t = x.text
        mobile_name.append(t)
    print(mobile_name)
    
    
    
    
    
```

```python
for x in url_src:
    names = name(x)
    print(names)
```

```python
mobile_name
```

```python
ratingss = []
def rating(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    ratings = soup.find_all("div", class_="a-row a-size-small")
    for x in ratings:
        r = x.text
        ratingss.append(r)
    print(ratingss)
    
    
```

```python
for x in url_src:
    rate = rating(x)
    print(rate)
```

```python
ratingss
```

```python
r  = [x.split(" ") for x in ratingss]
```

```python
f_r =  [x[0] for x in r]
```

```python
f_r
```

```python
prices = []
def m_price(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    price = soup.find_all("div" ,class_="a-row a-size-base a-color-base")
    for x in price:
        r = x.text.replace("₹"," ").replace(",","")
        prices.append(r)
    print(prices)
    
```

```python
for x in url_src:
    p = m_price(x)
    print(p)
```

```python
p_1 = [x.split(" ") for x in prices]
```

```python
f_p = [x[1] for x in p_1]
```

```python
f_p
```
