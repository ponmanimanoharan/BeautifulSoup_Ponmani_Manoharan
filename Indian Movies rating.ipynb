```python
soup = BeautifulSoup(page.content)
```

Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name,
IMDB rating, Year of release) and save it in form of a CSV file.

```python
detials = soup.find_all("span", class_= "lister-item-header")

```

```python
detials 
```

```python
data = [x.text.replace("\n", "*") for x in detials ]
```

```python
data
```

```python
data = [x.split("*") for x in data]
```

```python
data
```

```python
number = [x[1] for x in data]
movie_name = [x[3] for x in data]
year = [x[4] for x in data]
```

```python
print(number, movie_name, year)
```

```python
year = [x.replace("(", "").replace(")","").replace(" ","") for x in year ]
```

```python
year
```

```python
detials_four = soup.find_all(class_= "col-imdb-rating")
```

```python
detials_four
```

```python
ratings = [x.text.replace("\n","").replace(" ","") for x in detials_four]
ratings
```

```python
dict_header = dict(No = number, Name=movie_name, IMDB_rating=ratings, Year_of_release =year)
```

```python
import pandas as pd
IMBD_Rating = pd.DataFrame(dict([(keys, pd.Series(values)) for keys, values in dict_header.items()]))
```
