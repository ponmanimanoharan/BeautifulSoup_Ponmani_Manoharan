1. Write a python program to display all the header tags from
‘en.wikipedia.org/wiki/Main_Page’.

```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://en.wikipedia.org/wiki/Main_Page")
```

```python
page
```

```python
page.content 
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
heading_tags = soup.find_all("h1", class_="firstHeading")
```

```python
heading_tags
```

```python
heading_tags2 = soup.find_all("h2", class_="mp-h2")
```

```python
heading_tags2
```

```python

heading_tags3 = soup.find_all("h3", class_="vector-menu-heading")
```

```python
heading_tags3
```

```python
h1 = [x.text for x in heading_tags]
h2 = [x.text.replace("\xa0...", "") for x in heading_tags2]
h3 = [x.text.replace("\n", "") for x in heading_tags3]

```

```python
print(h1,h2,h3)
```

```python
dict_header = dict(header1 = h1, header2=h2, header3=h3)
```

```python
import pandas as pd
header_frame = pd.DataFrame(dict([(keys, pd.Series(values)) for keys, values in dict_header.items()]))
```
