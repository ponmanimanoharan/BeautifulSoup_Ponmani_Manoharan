Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You
have to scrape:
i) Top 10 ODI teams in men’s cricket along with the records for matches, points and
rating.
ii) Top 10 ODI Batsmen in men along with the records of their team and rating.
iii)Top 10 ODI bowlers along with the records of their team and rating.

```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
```

```python
soup
```

```python
def findval(tagname, classname):
    name = soup.find_all(tagname, class_=classname)
    return name
```

```python
top10oditeam  = findval("span", "u-hide-phablet")
```

```python
top10oditeam
```

```python
team = [x.text for x in top10oditeam]
```

```python
team
```

```python
def cutlist(newlist):
    newlist = newlist[0:10]
    return newlist
```

```python
team = cutlist(team)
print(team)
```

```python
top10odinumber = findval("td", "table-body__cell table-body__cell--position u-text-right")
```

```python
top10odinumber
```

```python
number =  [x.text for x in top10odinumber]
```

```python
number = cutlist(number)
print(number)
```

```python
number.insert(0,1)
```

```python
number.remove("11")
```

```python
number
```

```python
top10odimatches = findval("td", "table-body__cell u-center-text")
```

```python
matches = [x.text for x in top10odimatches]
```

```python
match = [x for x in matches if len(x) == 2]

```

```python
match.insert(0,"17")
```

```python
match = cutlist(match)
```

```python
print(match)
```

```python
point = [x.replace(",", "") for x in matches]
```

```python
points = [x for x in point if len(x) == 4]
```

```python
points.insert(0, '2054')
```

```python
points = cutlist(points)
```

```python
print(points)
```

```python
top10odiratings = findval("td", "table-body__cell u-text-right rating")
rating = [x.text for x in top10odiratings]
```

```python
rating.insert(0,"121")
```

```python
rating = cutlist(rating)
print(rating)
```

```python
dict_header = dict(Number = number, Team=team, Match=match, Points = points, Rating=rating)
```
