Write a python program to scrap book name, author name, genre and book review of
any 5 books from ‘www.bookpage.com’

```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://bookpage.com/reviews")
```

```python
soup = BeautifulSoup(page.content)
```

```python
books = soup.find_all("div", class_= "flex-article-content")
```

```python
books
```

```python
data = [x.text.replace("\n", "*") for x in books]
```

```python
data
```

```python
data = [x.split("*") for x in data]
```

```python
data
```

```python
book_name = [x[2] for x in data]
```

```python
print(book_name)
```

```python
author_name = [x[5] for x in data]
print(author_name)
```

```python
genre_1 = [x[8] for x in data]
print(genre_1)
```

```python
genre_2 = [x[10] for x in data]
print(genre_2)
```

```python
review = [x[13] for x in data]
print(review)
```

```python
dict_header = dict(name = book_name, author=author_name, genre_1 = genre_1,genre_2=genre_2, review = review)
```

```python
import pandas as pd
books = pd.DataFrame(dict([(keys, pd.Series(values)) for keys, values in dict_header.items()]))
```
